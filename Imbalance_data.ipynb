```python
# step-1:Data Alignment
import pandas as pd

#read sample data from Hospital dataset

df = pd.read_csv('https://raw.githubusercontent.com/Me-Rajdip/Hospital_Data/main/healthcare_dataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.columns
```

```python
df1 = df.drop(['Name','Doctor', 'Hospital', 'Insurance Provider'],axis=1)
```

```python
df1 = df[['Age', 'Gender', 'Blood Type', 'Medical Condition',
       'Date of Admission',
       'Billing Amount', 'Room Number', 'Admission Type', 'Discharge Date',
       'Medication', 'Test Results']]
df1
```

```python
df1 = pd.get_dummies(df1)
```

```python
df1.head()
```

```python
df1.Age.value_counts()
```

```python
# define y and X
df1.columns
```

```python
y = df1['Age']
X = df1[['Billing Amount', 'Room Number', 'Gender_Female', 'Gender_Male',
       'Blood Type_A+', 'Blood Type_A-', 'Blood Type_AB+', 'Blood Type_AB-',
       'Blood Type_B+', 'Blood Type_B-', 'Blood Type_O+', 'Blood Type_O-',
       'Discharge Date_2024-06-05', 'Discharge Date_2024-06-06',
       'Medication_Aspirin', 'Medication_Ibuprofen', 'Medication_Lipitor',
       'Medication_Paracetamol', 'Medication_Penicillin',
       'Test Results_Abnormal', 'Test Results_Inconclusive',
       'Test Results_Normal']]
```

```python
X
```

```python
y.value_counts()
```

```python
# imblearn undersampling
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()
X_rus, y_rus = rus.fit_resample(X,y)
```

```python
X.shape, X_rus.shape
```

```python
y.value_counts(), y_rus.value_counts()
```

```python
# imblearn over sampling
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_ros, y_ros = ros.fit_resample(X,y)
```

```python
X.shape, X_ros.shape
```

```python
y.value_counts(), y_ros.value_counts()
```

```python
# step train test split of original data
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y, random_state=30529)
```

```python
# create model
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
# train model original data
rfc.fit(X_train,y_train)
```

```python
# predict model original data
y_pred = rfc.predict(X_test)
```

```python
# evaluation original data
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# step train test split of  oversample data
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X_ros,y_ros, random_state=30529)
```

```python
# create model oversample data
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
# train model oversample data
rfc.fit(X_train,y_train)
```

```python
# predict model oversample data
y_pred = rfc.predict(X_test)
```

```python
# evaluation oversample data
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
